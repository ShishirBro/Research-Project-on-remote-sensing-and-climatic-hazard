```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
file_path = r'C:\Users\Acer\Desktop\all agricultural implications shit\SRYS master file.csv'
data = pd.read_csv(file_path)

# Calculate correlation coefficients for the specified SRYS columns with SWDIgl
correlation_coeffs = data[['SRYS_paddy', 'SRYS_maize', 'SRYS_millet', 'SRYS_wheat', 'SWDIgl']].corr().loc['SWDIgl', :]

# Visual representation with regression line and correlation values in titles
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

# SRYS_paddy
sns.regplot(data=data, x='SWDIgl', y='SRYS_paddy', ax=ax[0, 0], line_kws={"color": "r"}, scatter_kws={"color": "b"})
paddy_corr = correlation_coeffs['SRYS_paddy']
ax[0, 0].set_title(f'SRYS Paddy Yield vs SWDIgl\nCorrelation: {paddy_corr:.2f}', fontsize=11)
ax[0, 0].set_xlabel('', fontsize=11)

# SRYS_maize
sns.regplot(data=data, x='SWDIgl', y='SRYS_maize', ax=ax[0, 1], line_kws={"color": "r"}, scatter_kws={"color": "b"})
maize_corr = correlation_coeffs['SRYS_maize']
ax[0, 1].set_title(f'SRYS Maize Yield vs SWDIgl\nCorrelation: {maize_corr:.2f}', fontsize=11)
ax[0, 1].set_xlabel('', fontsize=11)
ax[0, 1].set_ylabel('', fontsize=11)

# SRYS_Millet
sns.regplot(data=data, x='SWDIgl', y='SRYS_millet', ax=ax[1, 0], line_kws={"color": "r"}, scatter_kws={"color": "b"})
millet_corr = correlation_coeffs['SRYS_millet']
ax[1, 0].set_title(f'SRYS Millet Yield vs SWDIgl\nCorrelation: {millet_corr:.2f}', fontsize=11)
ax[1, 0].set_xlabel('SWDIgl', fontsize=11)

# SRYS_wheat
sns.regplot(data=data, x='SWDIgl', y='SRYS_wheat', ax=ax[1, 1], line_kws={"color": "r"}, scatter_kws={"color": "b"})
wheat_corr = correlation_coeffs['SRYS_wheat']
ax[1, 1].set_title(f'SRYS Wheat Yield vs SWDIgl\nCorrelation: {wheat_corr:.2f}', fontsize=11)
ax[1, 1].set_xlabel('SWDIgl', fontsize=11)
ax[1, 1].set_ylabel('', fontsize=11)
# Set the x-labels only for the bottom row
ax[1, 0].set_xlabel('SWDIgl', fontsize=11)
ax[1, 1].set_xlabel('SWDIgl', fontsize=11)

# Set the y-labels for the first column
ax[0, 0].set_ylabel('SRYS Paddy', fontsize=11)
ax[1, 0].set_ylabel('SRYS Millet', fontsize=11)

# Set the tick parameters for both x and y ticks
for a in ax.flatten():
    a.tick_params(axis='both', which='major', labelsize=11)

# Adjust layout to make room for the main title
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.tight_layout()

# Save the figure
plt.savefig(r'C:\Users\Acer\Desktop\Manuscript edited plots\Crops vs SWDI.png', dpi=500, bbox_inches='tight')

# Show the plot
plt.show()

```
