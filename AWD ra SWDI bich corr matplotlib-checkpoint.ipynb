```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Since I cannot directly execute the code here due to the environment's limitations, I'll provide you with an updated version
# of the code that includes correlation values in the plot titles. This version assumes you've loaded `master_df` as shown previously.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Reload the master dataframe if not already loaded
master_df = pd.read_excel(r'C:\Users\Acer\Desktop\AWD\masterfile for cor of AWD and SWDI.xlsx')

def plot_awd_swdi_correlation_with_title(df, station_id, ax):
    awd_col = f'AWD_{station_id}'
    swdi_col = f'SWDI_{station_id}'
    
    # Filter out NaN values for accurate plotting and calculation
    valid_data = df[[awd_col, swdi_col]].dropna()
    x = valid_data[awd_col]
    y = valid_data[swdi_col]
    
    # Calculate best fit line and correlation
    m, b = np.polyfit(x, y, 1)
    correlation = np.corrcoef(x, y)[0, 1]
    
    # Plot
    ax.scatter(x, y, label=f'Station {station_id}', alpha=0.5)
    ax.plot(x, m*x + b, color='red', label=f'Best Fit Line')
    ax.set_xlabel('AWD',size=10)
    ax.set_ylabel('SWDI',size=10)
    ax.set_title(f'Station {station_id} (Correlation: {correlation:.2f})')
    ax.legend()

# Create a 2x2 subplot to compare all 4 stations
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
#fig.suptitle('AWD vs SWDI Correlation for All Stations')

# Plot for each station
station_ids = ['203', '307', '311', '402']
for station_id, ax in zip(station_ids, axs.flatten()):
    plot_awd_swdi_correlation_with_title(master_df, station_id, ax)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()



```
