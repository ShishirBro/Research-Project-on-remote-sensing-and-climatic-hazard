```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = r'C:\Users\Acer\Desktop\all agricultural implications shit\SRYS master file.csv'
```

```python
my_data1 = pd.read_csv(data1)
```

```python
my_data1
```

```python
my_data1['Year'] = pd.to_datetime(my_data1['Year'], format='%Y')
```

```python


# Assuming 'my_data1' is your DataFrame and 'Year' is stored as integer or datetime
plt.figure(figsize = (15, 5))
plt.plot(my_data1['Year'], my_data1['SRYS_paddy'], color='g', label ='paddy', marker = '*')
plt.plot(my_data1['Year'], my_data1['SRYS_maize'], color='b', label = 'Maize', marker = '*')
plt.plot(my_data1['Year'], my_data1['SRYS_millet'], color='r', label= 'Millet', marker = '*')
plt.plot(my_data1['Year'], my_data1['SRYS_wheat'], color='y', label= 'Wheat', marker = '*')
plt.plot(my_data1['Year'], my_data1['SWDIgl'], color='black', label= 'Agricultural Drought', marker = '*')

# Setting the x-axis to show every year explicitly
plt.xticks(my_data1['Year'][::3], size = 11)  # Rotation for better visibility if labels overlap
plt.yticks(size=11)
plt.legend()
plt.xlabel('Year', size=11)
plt.ylabel('SRYS', size=11)

# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig(r'C:\Users\Acer\Desktop\Manuscript edited plots\SRYS and drought.png', dpi=500, bbox_inches='tight')

# Show the plot
plt.show()
```
