```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Paths to your CSV files
data_paths = [
    r'C:\Karnali Drought assesment\SPEI\speitimeseries_203.csv',
    r'C:\Karnali Drought assesment\SPEI\speitimeseries_311.csv'
]
```

```python
station_numbers = ['203 (silgadi, Doti)', '311 (Simikot, Humla)']
```

```python
# List to store dataframes
dataframes = []
```

```python
# Read each CSV file into a dataframe and append to the list
for path in data_paths:
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])  # Ensure 'Date' column is datetime
    dataframes.append(df)
```

```python
dataframes[0]
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 10), dpi=500)
# Plotting data on each subplot
for idx, df in enumerate(dataframes):
    # Determine row location
    col = idx
    row = idx
    # Plot sfi6
    axes[0, col].plot(df['Date'], df['spei3'], color='b', marker = '*', markersize='4', label='spei3')
    axes[0, col].set_title(f'Station {station_numbers[idx]}', fontsize=11)
    axes[0, col].set_ylabel('spei3', fontsize=11)
    axes[0, col].xaxis.set_major_locator(mdates.YearLocator())
    axes[0, col].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    #if row == len(dataframes) - 1:  # Only set xlabel on the bottom row
        #axes[row, 0].set_xlabel('Date', fontsize=11)
    #axes[row, 0].legend(fontsize=11)
    
    # Plot sfi12
    axes[1, col].plot(df['Date'], df['spei6'], color='r', marker = '*', markersize='4', label='spei6')
    #axes[1, col].set_title(f'Station {station_numbers[idx]}', fontsize=11)
    axes[1, col].set_ylabel('spei6', fontsize=11)
    axes[1, col].xaxis.set_major_locator(mdates.YearLocator())
    axes[1, col].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    #if row == len(dataframes) - 1:  # Only set xlabel on the bottom row
        #axes[row, 1].set_xlabel('Date', fontsize=11)
   #axes[row, 1].legend(fontsize=11)
    
        # Plot sfi12
    axes[2, col].plot(df['Date'], df['spei12'], color='g', marker = '*', markersize='4', label='spei12')
    #axes[2, col].set_title(f'Station {station_numbers[idx]}', fontsize=11)
    axes[2, col].set_ylabel('spei12', fontsize=11)
    axes[2, col].xaxis.set_major_locator(mdates.YearLocator())
    axes[2, col].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    axes[2, col].set_xlabel('Date', fontsize=11)
    
    # Increase the spacing for the date ticks
    for ax in axes[row, :]:
        ax.tick_params(axis='both', which='major', labelsize=11)
# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig(r'C:\Users\Acer\Desktop\Manuscript edited plots\spei_plots.png', dpi=500, bbox_inches='tight')

# Show the plot
plt.show()
```
